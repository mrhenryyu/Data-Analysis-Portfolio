```python
#: means columns , row 0 all columns 
print(ufo.loc[0,: ])
```

```python
ufo.loc[0:10, :]
```

```python
#column selection 
ufo.loc[:, 'City']
```

```python
ufo.loc[0:2, 'City': 'State']
```

```python
ufo.head(3).drop('Time', axis = 1)
```

```python
#loc is also useful for boolean conditions 
ufo[ufo.City == 'Oakland']
```

```python
ufo.loc[ufo.City == 'Oakland', :]
```

```python
#using iloc 
# basically loc escept for INTEGER position 
ufo.iloc[:, 0:4]
#iloc is EXCLUSIVE of second number, inclusive of first number 
# it is like range(0,4). 
```

```python
ufo.iloc[0:3, :]
#loc is inclusive on both sides, iloc is not 
```

```python
ufo[['City', 'State']]
```

```python
ufo[0:2]
```

```python
#pandas ix 
# allows you to mix labels and integers, it is a blend of loc and iloc 

drinks = pd.read_csv('http://bit.ly/drinksbycountry', index_col = 'country')
```

```python
drinks.head()
```

```python
drinks.ix['Albania', 0]
```

```python
drinks.ix[1,'beer_servings']
```

```python
drinks.ix['Albania': 'Andorra', 0:2]
```

```python
drinks.ix['Canada',:]
```

```python
drinks.ix[0:2, 0:2]
```
