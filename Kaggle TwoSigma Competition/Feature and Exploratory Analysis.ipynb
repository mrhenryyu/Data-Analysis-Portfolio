```python
import pandas as pd 
import numpy as np 
import seaborn as sns 

import matplotlib.pyplot as plt 

%matplotlib inline 

# intersting color palate from SNS 
color = sns.color_palette()
```

Now we import the dataset and take a look at it. Lets see what the overall interests level are 

```python
train = pd.read_json(r'train.json')
train.head()
```

```python
# plot the value counts
interest = train.interest_level.value_counts()
plt.figure(figsize= (8, 4))
interest.plot(kind = 'bar', color = color[2])
#sns.barplot(interest.index, interest.values, color = color[2])
plt.xlabel('Interest Level')
plt.ylabel('Frequency')
```

So most have a low, with smaller medium and high. Their algorithm to determine the interest level must have set it that way. It's interesting how there's no square foot in their data at all, but maybe most people don't have that information on hand.

### Bathrooms 

Lets look at the average bathroom counts and its relation to interest level 

```python
bath_count = train['bathrooms'].value_counts().sort_index()
plt.figure(figsize = (10, 5))
sns.barplot(bath_count.index, bath_count.values, alpha = 0.7, color = color[3])
plt.xlabel('Number of bathrooms')
plt.ylabel('Frequency')
```

based on the symmetry of the data, it doesn't seem like bathroom number has a lot of correlation with interest. It's also correlated with price

## Price 



```python
# use a histogram 
plt.figure(figsize=(8,6))
plt.scatter(range(train.shape[0]), np.sort(train.price.values))
plt.xlabel('index', fontsize=12)
plt.ylabel('price', fontsize=12)
# set the xparameters 
plt.xlim(0,55000)
plt.ylim(-10000,150000)
```

```python
# get rid of the top percentile 
ulimit = np.percentile(train.price.values, 99)
train['price'].ix[train['price'] > ulimit] = ulimit
plt.figure(figsize = (8,6))
sns.distplot(train.price.values, bins = 50, kde = True)
```

```python
# dealing with longtiude and latitude 
```

```python
# use a distplot for longitude and latitude seperately 
```

```python
# latitude 
plt.figure(figsize = (8, 4))
sns.distplot(train.latitude.values, bins = 50)
# we can see limits really weird on there, we need to tie them up somehow 
# PS, since we are changing the data here, we can't use this for training 
```

```python
# percentile 
ulimit = np.percentile(train.latitude.values, 99)
llimit = np.percentile(train.latitude.values, 1)

train['latitude'].ix[train['latitude'] > ulimit] = ulimit
train['latitude'].ix[train['latitude'] < llimit] = llimit

plt.figure(figsize = (8, 6))
sns.distplot(train.latitude.values, bins = 30, kde = False)
```

```python
# repeat with longitude
ulimit = np.percentile(train.longitude.values, 99)
llimit = np.percentile(train.longitude.values, 1)

train['longitude'].ix[train['longitude'] > ulimit] = ulimit
train['longitude'].ix[train['longitude'] < llimit] = llimit

plt.figure(figsize = (8, 6))
sns.distplot(train.longitude.values, bins = 30, kde = False)
```

### Date Created 

Maybe that has something to do with the date it was created 

```python
train.created = pd.to_datetime(train['created'])
```

```python
train['date_created'] = train['created'].dt.date
```

```python
date_count = train['date_created'].value_counts().sort_index()
```

```python
plt.figure(figsize = (12, 5))

ax = plt.subplot(111)
ax.bar(date_count.index, date_count.values, alpha = 0.8)
ax.xaxis_date()
plt.xticks(rotation='vertical')
```

```python
# dates distribution of training 
# check if they are similar in distribution with testing 
test = pd.read_json('test.json')
```

```python
test["created"] = pd.to_datetime(test["created"])
test["date_created"] = test["created"].dt.date
cnt_srs = test['date_created'].value_counts()

plt.figure(figsize=(12,4))
ax = plt.subplot(111)
ax.bar(cnt_srs.index, cnt_srs.values, alpha=0.8)
ax.xaxis_date()
plt.xticks(rotation='vertical')
```

```python
# lets look at hour listing trend (minute is useless, so is hour, but why not )
hour_created = train['created'].dt.hour
hour_count = hour_created.value_counts().sort_index()
```

```python
plt.figure(figsize=(12,5))

ax = plt.subplot(111)
ax.bar(hour_count.index, hour_count.values, alpha = 0.8, color = color[4])
```

That's interesting, most are conducted in the morning


## Display Address

```python
plt.figure(figsize = (12,5))
train.display_address.value_counts().plot(kind = 'hist', bins = 50, xlim = (0,400), log = True, color = color[1])
plt.xlabel('Display Address Frequency')
```

```python
# also look at num_photos 
train.photos
# we can't get the quality of the photos yet, though we can use the number of photos 
train['num_photos'] = train.photos.apply(len)
```

```python
plt.figure(figsize=(12,6))
cnt_srs = train.num_photos.value_counts()

sns.barplot(cnt_srs.index, cnt_srs.values, alpha = 0.8)
plt.xlabel('Number of Photos')
```

```python
# do stacked bar with number of photos 
plt.figure(figsize=(20, 20))
ax = train.groupby('num_photos').interest_level.value_counts().sort_index().unstack().plot( kind = 'bar', stacked = True, xlim = (0,20), figsize = (12, 6))
ax.set_xlim(-0.5,20.5)
```

```python
# all of them have similar amount of photos, however, it seems like the ones with high interest have very little 0 photos 
```

```python


from wordcloud import WordCloud
```

```python

text = ''
text_da = ''
text_desc = ''
for ind, row in train.iterrows():
    for feature in row['features']:
        text = " ".join([text, "_".join(feature.strip().split(" "))])
    text_da = " ".join([text_da,"_".join(row['display_address'].strip().split(" "))])
    #text_desc = " ".join([text_desc, row['description']])
text = text.strip()
text_da = text_da.strip()
text_desc = text_desc.strip()

plt.figure(figsize=(12,6))
wordcloud = WordCloud(background_color='white', width=600, height=300, max_font_size=50, max_words=40).generate(text)
wordcloud.recolor(random_state=0)
plt.imshow(wordcloud)
plt.title("Wordcloud for features", fontsize=30)
plt.axis("off")
plt.show()

# wordcloud for display address
plt.figure(figsize=(12,6))
wordcloud = WordCloud(background_color='white', width=600, height=300, max_font_size=50, max_words=40).generate(text_da)
wordcloud.recolor(random_state=0)
plt.imshow(wordcloud)
plt.title("Wordcloud for Display Address", fontsize=30)
plt.axis("off")
plt.show()
```
