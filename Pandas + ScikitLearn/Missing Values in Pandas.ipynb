```python
#missing values in Pandas
import pandas as pd 
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
```

```python
ufo.tail()
```

```python
#NaN is not a number, it means a missing value 
ufo.isnull().tail()
```

```python
ufo.notnull().tail()
```

```python
#use isnull and notnull to get the amount of null 
#number of missing values in each of the columns. 
# sum does axis = 0 by default 
ufo.isnull().sum()
```

```python
pd.Series([True, False, True]).sum()
```

```python
#isnull is a series method 
ufo[ufo.City.isnull()]
#see a portion of the dataframe, with no city 
```

```python
# how to drop missing values 
ufo.shape
```

```python
ufo.dropna(how = 'any').shape
#drop a row if any of its values are missing 
#inplace = False 
```

```python
ufo.dropna(how = 'all').shape
#only drop a row if ALL of it is missing 
```

```python
ufo.dropna(subset = ['City', 'Shape Reported'], how = 'all').shape
```

```python
#filling missing values
ufo['Shape Reported'].value_counts(dropna = False)
```

```python
ufo['Shape Reported'].fillna(value = 'VARIOUS', inplace = True)
```

```python
ufo['Shape Reported'].value_counts()
```
